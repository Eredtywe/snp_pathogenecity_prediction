```python
import pandas as pd
import numpy as np
```

```python
snp_id = []
pos = []
ref = []
alt = []
sig = []
```

```python
sift_path = 'pred_output\HAP1_sift_pred_output.txt'
poly_path = 'pred_output\HAP1_poly_pred_output.txt'
phd_path  = 'pred_output\HAP1_phd_pred_output.txt'
```

```python
with open(sift_path, 'r') as sift_file:
    lines = sift_file.readlines()
```

```python
for line in lines[1:]:
    line_split = line.strip().split('\t')
    snp_id.append(line_split[0])
    pos.append(line_split[2])
    sig.append(line_split[11])
    temp = line_split[4]
    temp_split = temp.split('/')
    ref.append(temp_split[0])
    alt.append(temp_split[1])
```

```python
sift_data = {'snp_id' : snp_id,
        'pos' : pos,
        'ref' : ref,
        'alt' : alt,
        'sig' : sig}
```

```python
sift_df = pd.DataFrame(sift_data)
```

```python
sift_df
```

```python
with open(poly_path, 'r') as poly_file:
    lines = poly_file.readlines()
```

```python
snp_id = []
pos = []
ref = []
alt = []
sig = []
```

```python
for line in lines[1:]:
    line_split = line.strip().split('\t')
    snp_id.append(line_split[0])
    pos.append(line_split[2])
    sig.append(line_split[11])
    temp = line_split[4]
    temp_split = temp.split('/')
    ref.append(temp_split[0])
    alt.append(temp_split[1])
```

```python
poly_data = {'snp_id' : snp_id,
        'pos' : pos,
        'ref' : ref,
        'alt' : alt,
        'sig' : sig}
```

```python
poly_df = pd.DataFrame(poly_data)
```

```python
poly_df
```

```python
pos = []
ref = []
alt = []
sig = []
```

```python
with open(phd_path, 'r') as phd_file:
    lines = phd_file.readlines()
lines = lines[13:4963]
```

```python
for line in lines[1:]:
    line_split = line.strip().split('\t')
    pos.append(line_split[1])
    ref.append(line_split[3])
    alt.append(line_split[4])
    sig.append(line_split[6])
```

```python
phd_data = {'pos' : pos,
        'ref' : ref,
        'alt' : alt,
        'sig' : sig}
```

```python
phd_df = pd.DataFrame(phd_data)
```

```python
phd_df
```

```python
compare_columns = ['snp_id', 'pos', 'ref', 'alt']
sift_poly = pd.merge(sift_df, poly_df, on=compare_columns, how='outer')  
```

```python
sift_poly
```

```python
sift_poly = sift_poly.drop_duplicates()
```

```python
sift_poly
```

```python
compare_columns = ['pos', 'ref', 'alt']
sift_poly_phd = pd.merge(sift_poly, phd_df, on=compare_columns, how='outer')  
```

```python
sift_poly_phd
```

```python
sift_poly_phd['sig_x'] = sift_poly_phd['sig_x'].apply(lambda x: 1 if x == 'Deleterious' else 0)
sift_poly_phd['sig_y'] = sift_poly_phd['sig_y'].apply(lambda x: 0 if x == 'Benign' else 1)
sift_poly_phd['sig'] = sift_poly_phd['sig'].apply(lambda x: 0 if x == 'Benign' else 1)
```

```python
sift_poly_phd
```

```python
sift_poly_phd['total_score'] = sum([sift_poly_phd['sig_x'], sift_poly_phd['sig_y'], sift_poly_phd['sig']])
```

```python
sift_poly_phd
```

```python
HAP1_sift_counts = sift_poly_phd['sig_x'].value_counts()
HAP1_poly_counts = sift_poly_phd['sig_y'].value_counts()
HAP1_phd_counts = sift_poly_phd['sig'].value_counts()
HAP1_total_counts = sift_poly_phd['total_score'].value_counts()
```
