```python
import pandas as pd
import numpy as np
```

```python
sift_df = pd.read_csv('pred_output\HTT_sift_pred_output_mod.csv')
poly_df = pd.read_csv('pred_output\HTT_poly_pred_output_mod.csv')
phd_path = 'pred_output\HTT_phd_pred_output.txt'
```

```python
pos = []
ref = []
alt = []
sig = []
```

```python
with open(phd_path, 'r') as phd_file:
    lines = phd_file.readlines()
lines = lines[13:4963]
```

```python
for line in lines[1:]:
    line_split = line.strip().split('\t')
    pos.append(line_split[1])
    ref.append(line_split[3])
    alt.append(line_split[4])
    sig.append(line_split[6])
```

```python
phd_data = {'pos' : pos,
        'ref' : ref,
        'alt' : alt,
        'sig' : sig}
```

```python
phd_df = pd.DataFrame(phd_data)
```

```python
phd_df
```

```python
compare_columns = ['snp_id', 'pos', 'ref', 'alt']
sift_poly = pd.merge(sift_df, poly_df, on=compare_columns, how='outer')  
```

```python
sift_poly
```

```python
sift_poly = sift_poly.drop_duplicates()
```

```python
sift_poly
```

```python
sift_poly.dtypes
```

```python
phd_df.dtypes
```

```python
phd_df['pos'] = phd_df['pos'].astype(int)
```

```python
compare_columns = ['pos', 'ref', 'alt']
sift_poly_phd = pd.merge(sift_poly, phd_df, on=compare_columns, how='outer')  
```

```python
sift_poly_phd
```

```python
sift_poly_phd['sig_x'] = sift_poly_phd['sig_x'].apply(lambda x: 1 if x == 'Deleterious' else 0)
sift_poly_phd['sig_y'] = sift_poly_phd['sig_y'].apply(lambda x: 0 if x == 'Benign' else 1)
sift_poly_phd['sig'] = sift_poly_phd['sig'].apply(lambda x: 0 if x == 'Benign' else 1)
```

```python
sift_poly_phd
```

```python
sift_poly_phd['total_score'] = sum([sift_poly_phd['sig_x'], sift_poly_phd['sig_y'], sift_poly_phd['sig']])
```

```python
sift_poly_phd
```

```python
HTT_sift_counts = sift_poly_phd['sig_x'].value_counts()
HTT_poly_counts = sift_poly_phd['sig_y'].value_counts()
HTT_phd_counts = sift_poly_phd['sig'].value_counts()
HTT_total_counts = sift_poly_phd['total_score'].value_counts()
```
