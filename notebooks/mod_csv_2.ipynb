```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('interim_data\HAP1_snps_2.csv')
```

```python
df
```

```python
df = df[['snp_id', 'chr', 'pos', 'ref', 'alt_1', 'alt_2', 'alt_3']]
```

```python
df
```

```python
alt_all = df[['alt_1', 'alt_2', 'alt_3']].values.flatten()
```

```python
alt_all
```

```python
df = df.drop('alt_1', axis=1)
df = df.drop('alt_2', axis=1)
df = df.drop('alt_3', axis=1)
```

```python
sno = np.arange(1, len(df['pos']) + 1)
df['sno'] = sno
```

```python
df = df[['sno', 'snp_id', 'chr', 'pos', 'ref']]
```

```python
df
```

```python
df = pd.concat([df, df, df], ignore_index=True)
```

```python
df
```

```python
df = df.sort_values('sno', ascending=True)
```

```python
df
```

```python
df['alt'] = alt_all
```

```python
df
```

```python
df[0:40]
```

```python
sift_poly_csv = df
phd_snp_csv = df
```

```python
sift_poly_csv = sift_poly_csv[['snp_id']]
sift_poly_csv = sift_poly_csv.reset_index(drop=True)
```

```python
sift_poly_csv['dbsnp'] = 'dbsnp'
```

```python
sift_poly_csv = sift_poly_csv[['dbsnp', 'snp_id']]
```

```python
sift_poly_csv
```

```python
sift_poly_csv = sift_poly_csv.drop_duplicates()
```

```python
sift_poly_csv
```

```python
phd_snp_csv
```

```python
phd_snp_csv = phd_snp_csv.reset_index(drop=True)
phd_snp_csv = phd_snp_csv.drop('sno', axis=1)
phd_snp_csv = phd_snp_csv.drop('snp_id', axis=1)
```

```python
phd_snp_csv
```
