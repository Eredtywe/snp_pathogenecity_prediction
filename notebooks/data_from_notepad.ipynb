```python
import numpy as np
import pandas as pd
```

```python
file_path = 'HTT_snps.txt'
```

```python
snv_data = np.empty((70000,), dtype=f'S{30}')
snp_id = np.empty((70000,), dtype=f'S{30}')
chromosome = np.empty((70000,), dtype=f'S{30}')
position = np.empty((70000,), dtype=f'S{30}')
signature = np.empty((70000,), dtype=f'S{30}')
remove = np.empty((70000,), dtype=int)
num = np.empty((70000,), dtype=int)
```

```python
with open(file_path, 'r') as file:
    lines = file.readlines()
```

```python
i = 0
k = 1
```

```python
for line in lines:
    words = line.split()
    
    if words[0] == str(k) + '.':
        snp_id[i] = words[1]
        k += 1
        i += 1

    if line[0:3] == 'SNV':
        temp = line.split(':')
        snv_data[i-1] = temp[1].rstrip('\n')
        remove[i-1] = 0

    if line[0:6] == 'DELINS':
        temp = line.split(':')
        snv_data[i-1] = temp[1].rstrip('\n')
        remove[i-1] = 1
    
    if words[0] == 'Chromosome:':
        temp = words[1].split(":")
        chromosome[i-1] = temp[0]
        position[i-1] = temp[1]
    
    if words[0] == 'Clinical':
        signature[i-1] = words[2]
```

```python
for i in range(len(snp_id)):
    num[i] = i+1
```

```python
data = {'sn': num,
        'snp_id': snp_id,
        'snv_data': snv_data,
        'chromosome': chromosome,
        'position': position,
        'signature': signature,
        'remove': remove}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['snp_id'] = df['snp_id'].apply(lambda x: x.decode('utf-8').strip('b\'\''))
df['snv_data'] = df['snv_data'].apply(lambda x: x.decode('utf-8').strip('b\'\''))
df['chromosome'] = df['chromosome'].apply(lambda x: x.decode('utf-8').strip('b\'\''))
df['position'] = df['position'].apply(lambda x: x.decode('utf-8').strip('b\'\''))
df['signature'] = df['signature'].apply(lambda x: x.decode('utf-8').strip('b\'\''))
```

```python
df['signature'] = df['signature'].replace('enign', 'benign')
```

```python
df
```

```python
df = df[remove == 0 ]
```

```python
df[0:20]
```

```python
df = df.applymap(str)
df = df.drop('remove', axis=1)
df = df.dropna(axis=0, how='all')
df = df[df['snp_id'].str.startswith('r')]

```
